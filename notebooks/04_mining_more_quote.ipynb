```python
import pandas as pd
import json
import itertools
```

```python
df=pd.read_csv('/var/DSTeam/locnt2/personal/quotes-tagging/remain_quote.csv')
```

```python
cate_roots= json.load(open('/var/DSTeam/locnt2/personal/quotes-tagging/data/category.json'))
```

```python
cate_pd_total=pd.DataFrame()
for big_cat in cate_roots.keys():
    print(big_cat)
    bigcate=[big_cat]
    cates=[cate_roots[big_cat][i]['tag'] for i in range(len(cate_roots[big_cat]))]
    pairs=list(itertools.product(bigcate,cates))
    cate_pd=pd.DataFrame(pairs,columns=['big_cate','tag'])
    
    cate_pd_total=pd.concat([cate_pd_total,cate_pd])
```

```python
df.assign(
    tag=lambda x:x.tags.str.lower()
).assign(
    tag_origin=lambda x:x.tag.str.split(',')
).explode('tag_origin')[['quote','author','tag_origin']].sample(10).merge(
    cate_pd_total,
    how='cross'
)
```
